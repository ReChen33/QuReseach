# LTRA model

We should be able to model the wiring without a shield and with shield floating using the SPICE LTRA model.  Here's what the LTSpice model looks like:

Where Rterm=1G simulates the twisted pair being open at the end furthest from the VNA.  Rperlen is the roundtrip resistance per unit length in Ohm/meter, Lperlen is the inductance per unit length in Henry/meter, and Cperlen is the capacitance per unit length in Farad/m.  length is the length in meters.  Here's a simple function using PyLTSpice to compute the S11 using LTSpice for a given set of these parameters:

```python
def ltspice_ltra_s11(Rperlen=72.2, Lperlen=711e-9, Cperlen=58e-12, length=1.232):
    LTC = SimRunner(output_folder='./temp')
    LTC.create_netlist('sparam_ltra.asc')
    netlist = SpiceEditor('sparam_ltra.net')
    
    netlist.set_parameters(Rperlen=Rperlen)
    netlist.set_parameters(Lperlen=Lperlen)
    netlist.set_parameters(Cperlen=Cperlen)
    netlist.set_parameters(length=length)

    raw, log = LTC.run_now(netlist, run_filename=f'sparam_ltra.net')
    LTC.wait_completion()
    rr=RawRead(raw)
    s11=20.*np.log10(np.abs(rr.get_trace('S11(v1)')))
    f=rr.get_trace('frequency')
    return np.array(f,dtype=float),s11
```

Fitting using s11_fitting/ltra_fit.py, the LTRA model describes the S11 data for the TWP only and floating shield twisted pair pretty well, at least <100 MHz:

```python
ltra_floating_shield_f,ltra_floating_shield_s11=ltspice_ltra_s11(Rperlen=72.2,Lperlen=711e-9,Cperlen=58e-12,length=1.232)
ltra_no_shield_f,ltra_no_shield_s11=ltspice_ltra_s11(Rperlen=72.2,Lperlen=711e-9,Cperlen=30.1e-12,length=1.25)
```

```python
# Plot data
plt.semilogx(no_shield_fdata,no_shield_s11data,lw=3,label='No shield')
plt.semilogx(floating_shield_fdata,floating_shield_s11data,lw=3,label='Floating shield')

# Plot LTRA model
plt.semilogx(ltra_floating_shield_f,ltra_floating_shield_s11,'r--', label='LTRA fit floating shield')
plt.semilogx(ltra_no_shield_f,ltra_no_shield_s11,'m--', label='LTRA fit no shield')

# Higher frequencies will be hard to model due to the skin effect
plt.xlim(1e4,1e8)

#Legend
plt.legend(loc='lower left')
```

So that's pretty neat.  Agrees well, out to nearly 100 MHz!

# Do the LTRA fitted parameters to floating and no shield data make sense?

Based just on cable geometry and materials, expect the capacitance of the bare twisted pair to be 32.4 pF/m and LTRA model fits 30.1 pF/m.  The capacitance with the shield floating is related to the interpair and wire-shield capacitance like this (from https://www.nationalwire.com/support_basicinfo.php):

So our two LTRA fits would imply the shield-wire capacitance is (58 pF/m-30 pF/m)*2 = 56 pF/m.

Does that make sense?  Assuming shield is a circle, the inner shield diameter is 28 mils.  Ignoring the other wire and pretending it's a coax, the center conductor is 5 mils in diameter.  Roughly 1/4 of the inner diameter of the shield is FEP, the rest presumably air or vacuum, so the effective dielectric constant somewhere between ~1.25 and 2.01.  That range gives a coaxial capacitance per length of between 40-65 pF/m, which we're right in the middle of, so maybe this makes sense.

# Lumped element modeling to try and capture how the shield being grounded changes the circuit

To model how the shield being connected changes the circuit, need to move beyond the LTRA model.  Let's try to simulate the transmission line using a discrete number of lumped element circuits.  First let's check that for the same parameters, this model agrees with the LTRA model.  Let's try the floating shield case.  Here's what the lumped element model looks like for the shield floating case:

So now it's a balanced circuit, and there's a third shield conductor.  There are also 4 new parameters ; Rshield, Lsperlen, Csperlen, and Rsperlen ; these are the resistance between the shield and ground at the source, the shield resistance, shield inductance, and the capacitance between each wire and the shield (which is assumed to be the same for both wires).  First thing to check is if this agrees with the LTRA model.  Here's a simple function for running this simulation (where de256 stands for 256 discrete elements):

```python
def ltspice_de256_s11(Rperlen=72.2, Lperlen=711e-9, Cperlen=30.1e-12, Lsperlen=0, Csperlen=56e-12, Rsperlen=5.93, Rshield=1e9, Lpar=0, length=1.232):
    LTC = SimRunner(output_folder='./temp')
    LTC.create_netlist('sparam_shieldedbalancedlumpedline256_shieldCRL_shieldfloating.asc')
    netlist = SpiceEditor('sparam_shieldedbalancedlumpedline256_shieldCRL_shieldfloating.net')
    
    netlist.set_parameters(Rperlen=Rperlen)
    netlist.set_parameters(Lperlen=Lperlen)
    netlist.set_parameters(Cperlen=Cperlen)
    netlist.set_parameters(Rshield=Rshield)
    netlist.set_parameters(Rsperlen=Rsperlen)
    netlist.set_parameters(Lsperlen=Lsperlen)
    netlist.set_parameters(Csperlen=Csperlen)
    netlist.set_parameters(Lpar=Lpar)
    netlist.set_parameters(length=length)

    raw, log = LTC.run_now(netlist, run_filename=f'sparam_shieldedbalancedlumpedline256_shieldCRL_shieldfloating.net')
    LTC.wait_completion()
    rr=RawRead(raw)
    s11=20.*np.log10(np.abs(rr.get_trace('S11(v1)')))
    f=rr.get_trace('frequency')
    return np.array(f,dtype=float),s11
```

Let's run them side by side;

```python
ltra_floating_shield_f,ltra_floating_shield_s11=ltspice_ltra_s11(Rperlen=72.2,Lperlen=711e-9,Cperlen=58e-12,length=1.232)
de256_floating_shield_f,de256_floating_shield_s11=ltspice_de256_s11(Rperlen=72.2,Lperlen=711e-9,Cperlen=30.1e-12,Rshield=1e9,Rsperlen=5.93,Lsperlen=0,Csperlen=56e-12,length=1.232)
```

The lumped element model takes a *lot* longer to simulate.  Here I've set the shield self inductance to zero and the resistance to the measured DC value for the cable (if the shield is floating and this resistance is small, the simulation takes a significantly longer time, for some reason).  Do they agree?

```python
plt.semilogx(ltra_floating_shield_f,ltra_floating_shield_s11,'b', label='LTRA model')
plt.semilogx(de256_floating_shield_f,de256_floating_shield_s11,'c--', label='Discrete 256 element 3-conductor model')

# Higher frequencies will be hard to model due to the skin effect
plt.xlim(1e4,1e8)
plt.xlabel('Frequency (Hz)')
plt.ylabel('S11 (dB)')

#Legend
plt.legend(loc='lower left')
```

Yes, these agree.  Let's try tying the shield to ground at port 1 now and comparing to the measurement.  The shield connection to ground must be finite or the simulation will crash.

```python
de256_floating_shield_f,de256_floating_shield_s11=ltspice_de256_s11(Rperlen=72.2,Lperlen=711e-9,Cperlen=30.1e-12,Rshield=1.e-9,Rsperlen=5.93,Lsperlen=0,Csperlen=56e-12,length=1.232)
```

```python
plt.semilogx(conn_shield_fdata,conn_shield_s11data,label='Connected shield data')
plt.semilogx(de256_floating_shield_f,de256_floating_shield_s11,'c--', label='Discrete 256 element 3-conductor model')

# Higher frequencies will be hard to model due to the skin effect
plt.xlim(1e4,1e8)
plt.xlabel('Frequency (Hz)')
plt.ylabel('S11 (dB)')

#Legend
plt.legend(loc='lower left')
```

Interesting ; qualitative frequency behavior is spot on, but agreement is not great.  We have one additional parameter in this model, the shield self-inductance.  How does it change the S11?  This cable used to take this data was also slightly shorter, ~1.196m instead of 1.232m.

```python
dvsLs={}
# nH
for Ls in [0,25,50,75,100]:
    dvsLs[Ls]={}
    dvsLs[Ls]['f'],dvsLs[Ls]['s11']=ltspice_de256_s11(Rperlen=72.2,Lperlen=711e-9,Cperlen=30.1e-12,Rshield=1.e-9,Rsperlen=5.93,Lsperlen=Ls*1e-9,Csperlen=56e-12,length=1.196)
```

```python
plt.semilogx(conn_shield_fdata,conn_shield_s11data,label='Connected shield data')
for Ls in [0,25,50,75,100]:
    plt.semilogx(dvsLs[Ls]['f'],dvsLs[Ls]['s11'],'--', label=f'Discrete 256 element 3-conductor model Ls={Ls} nH')

# Higher frequencies will be hard to model due to the skin effect
plt.xlim(1e4,1e8)
plt.ylim(-30,3)
plt.xlabel('Frequency (Hz)')
plt.ylabel('S11 (dB)')

#Legend
plt.legend(loc='lower left')
```

That's cool.  50 nH/m looks closest;

```python
plt.semilogx(conn_shield_fdata,conn_shield_s11data,label='Connected shield data')
for Ls in [50]:
    plt.semilogx(dvsLs[Ls]['f'],dvsLs[Ls]['s11'],'--', label=f'Discrete 256 element 3-conductor model Ls={Ls} nH')

# Higher frequencies will be hard to model due to the skin effect
plt.xlim(1e4,1e8)
plt.ylim(-30,3)
plt.xlabel('Frequency (Hz)')
plt.ylabel('S11 (dB)')

#Legend
plt.legend(loc='lower left')
```

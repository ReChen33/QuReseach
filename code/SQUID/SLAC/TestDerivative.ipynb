```python
# open the file

import pickle
import gzip

data_file = 'results_1717115674_c8.pklz'

with gzip.open(data_file, "rb") as file:
    data = pickle.load(file)

```

```python
import matplotlib.pylab as plt
import numpy as np

x, y = data[8]['sa_fb_uA'], data[8]['sa_out_mV']

plt.plot(x, y, "b-")
plt.xlabel('SSA Feedback Current (uA)')
plt.ylabel('SSA Output Voltage (mV)')
plt.show()


```

```python
# unified Order of magnitude for unit uA and mV
x_mA = x * 1e-3 # change uA to mA
#y1 = y * 1e3 # mV to uV

# calculate the derivative
def getDerivative(x, y):
    dy_forward = np.diff(y) / np.diff(x)
    half_diff = np.diff(x) / 2
    x_mid = x[:-1] + half_diff
    return dy_forward, x_mid

dy, dx = getDerivative(x_mA, y)
ddy, ddx = getDerivative(dx, dy)

print(len(dy), len(ddy))
```

```python
plt.plot(x_mA, y, "b-")
plt.xlabel('SSA Feedback Current (mA)')
plt.ylabel('SSA Output Voltage (mV)')
plt.show()

plt.plot(dx, dy, "r-")
plt.show()

plt.plot(ddx, ddy, "g-")
plt.show()
```
